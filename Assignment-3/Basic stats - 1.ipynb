```python
import pandas as pd
import numpy as np
import statistics 
```

```python
# Load the dataset.
df = pd.read_csv("sales_data_with_discounts.csv")
df.head()
```

```python
# Identifing numerical columns in dataset.
df.select_dtypes(include = 'number').columns
```

```python
# Calculate the mean, median, mode, and standard deviation for these columns.
```

```python
# 1. Volume column in data set
print("Mean for Volume column:", np.round(df['Volume'].mean(), 2))
print("Median for Volume column:", np.round(df['Volume'].median(), 2))
print("Mode for Volume column:", np.round(statistics.mode(df['Volume']), 2))
print("Standard deviation for Volume column:", np.round(df['Volume'].std(), 2))
```

```python
# INTERPRETATION
'''
    The average volume is 5.07, with a median of 4 and a mode of 3, indicating that most transactions involve smaller volumes.
    Since the mean is higher than the median, the distribution of volume is right-skewed, 
    suggesting the presence of a few large-volume transactions. 
    The standard deviation of 4.23 shows high variability in the volume data, 
    meaning the number of units sold varies significantly across transactions.
'''
```

```python
# 2. Avg Price column in data set
print("Mean for Avg Price column:", np.round(df['Avg Price'].mean(), 2))
print("Median for Avg Price column:", np.round(df['Avg Price'].median(), 2))
print("Mode for Avg Price column:", np.round(statistics.mode(df['Avg Price']), 2))
print("Standard deviation for Avg Price column:", np.round(df['Avg Price'].std(), 2))
```

```python
# INTERPRETATION
'''
    The average price is around ₹10,453, but the median is much lower at ₹1,450, 
    showing that the data is highly right-skewed — a few very expensive items are increasing the average. 
    The high standard deviation indicates large variation in prices, 
    meaning the dataset contains both low- and high-priced products.
'''
```

```python
# 3. Total Sales Value column in data set
print("Mean for Total Sales Value column:", np.round(df['Total Sales Value'].mean(), 2))
print("Median for Total Sales Value column:", np.round(df['Total Sales Value'].median(), 2))
print("Mode for Total Sales Value column:", np.round(statistics.mode(df['Total Sales Value']), 2))
print("Standard deviation for Total Sales Value column:", np.round(df['Total Sales Value'].std(), 2))
```

```python
# INTERPRETATION
'''
    The average total sales value is ₹33,812.84, but most sales are much lower (median ₹5,700). 
    This again suggests a right-skewed distribution where a few very high sales are pulling the mean upward. 
    The large standard deviation reflects significant variability among sales amounts, 
    indicating a mix of small and large sales transactions.
'''
```

```python
# 4. Discount Rate (%) column in data set
print("Mean for Discount Rate (%) column:", np.round(df['Discount Rate (%)'].mean(), 2))
print("Median for Discount Rate (%) column:", np.round(df['Discount Rate (%)'].median(), 2))
print("Mode for Discount Rate (%) column:", np.round(statistics.mode(df['Discount Rate (%)']), 2))
print("Standard deviation for Discount Rate (%) column:", np.round(df['Discount Rate (%)'].std(), 2))
```

```python
# INTERPRETATION
'''
    The average discount rate offered is around 15%, 
    and the distribution is fairly balanced since the mean and median are close. 
    The small standard deviation (4.22) suggests that most discounts fall within a narrow range, 
    showing consistent discounting practices across transactions.
'''
```

```python
# 5. Discount Amount column in data set
print("Mean for Discount Amount column:", np.round(df['Discount Amount'].mean(), 2))
print("Median for Discount Amount column:", np.round(df['Discount Amount'].median(), 2))
print("Mode for Discount Amount column:", np.round(statistics.mode(df['Discount Amount']), 2))
print("Standard deviation for Discount Amount column:", np.round(df['Discount Amount'].std(), 2))
```

```python
# INTERPRETATION
'''
    The mean discount amount is ₹3,346.5, 
    but the median is only ₹988.93 — showing that a few very large discounts are raising the average. 
    The distribution is right-skewed, 
    and the high standard deviation indicates wide variation in discount amounts across sales.
'''
```

```python
# 6. Net Sales Value column in data set
print("Mean for Net Sales Value column:", np.round(df['Net Sales Value'].mean(), 2))
print("Median for Net Sales Value column:", np.round(df['Net Sales Value'].median(), 2))
print("Mode for Net Sales Value column:", np.round(statistics.mode(df['Net Sales Value']), 2))
print("Standard deviation for Net Sales Value column:", np.round(df['Net Sales Value'].std(), 2))
```

```python
# INTERPRETATION
'''
    The mean net sales value is ₹30,466, while the median is only ₹4,678, 
    again suggesting a few very large sales are inflating the average. 
    The mode being extremely high (₹160,346.5) shows outliers or unusual sales records. 
    The large standard deviation confirms high variability in sales performance.
'''
```
